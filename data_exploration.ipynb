## Visualizing more than 2 features

```python
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
# what is the relationship between total bill and tip based on sex and day
# this shows the relationship between 4 features

sns.scatterplot(x='total_bill', y='tip', hue='sex', style='day', size='size', palette='colorblind', data=df_tips);

# move legend out of the graph
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# different plots showing in one figure
sns.relplot(
    x='total_bill',
    y='tip',
    hue='sex',
    style='day',
    size='size',
    row='time',
    palette='colorblind',
    data=df_tips
    );
```

```python
sns.catplot(
    x='time',
    hue='day',
    col='smoker',
    kind='count',
    data=df_tips
    );
```

```python
def show_outliers_for_each_feature(df):
  for col in df.columns:
    sns.boxplot(data=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()

show_outliers_for_each_feature(df_tips)
```

```python
def grab_outliers(col):
  q1 = col.quantile(0.25)
  q3 = col.quantile(0.75)
  iqr = q3 - q1

  outliers = col[((col < (q1-1.5*iqr)) | (col > (q3+iqr*1.5)))]
  return outliers

for col in df_tips[['tip', 'total_bill']]:
  outliers = grab_outliers(df_tips[col])
  print(f'Number of outliers in {col}: {len(outliers)}')


# outliers_tips = grab_outliers(df_tips['tip'])
# outliers_tips
# print('Maximum Tips: ', outliers_tips.max())
# print('Minimum Tips: ', outliers_tips.min())
# print('Number of tip outliers: ', len(outliers_tips))
```

```python
# drop all outliers and print total outliers
def drop_outliers(dataset, outliers, col):
  for outlier in outliers:
    print('Outlier: ', outlier)
    for index, row in dataset.iterrows():
      if outlier == row[col]:
        # print('Row: ', row)
        dataset = dataset.drop(index, axis=0)
  return dataset

df_tips = drop_outliers(df_tips, tips_outliers, 'tip')
print(f'Total outliers: {len(tips_outliers)}')
tips_outliers = grab_outliers(df_tips['tip'])
print(f'Total outliers: {len(tips_outliers)}')
# Removing the outliers creates a new outlier this subsequent run removes that new outlier
df_tips = drop_outliers(df_tips, tips_outliers, 'tip')
tips_outliers = grab_outliers(df_tips['tip'])
print(f'Total outliers: {len(tips_outliers)}')
```
